```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete
```

Calculate the total number of students## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_column = school_data.iloc[:, 1] 
# Count the number of unique schools
total_unique_schools = school_column.nunique()
```

```python
# Calculate the total number of students
student_column = student_data.iloc[:, 1]
total_unique_student = student_column.count()
```

```python
# Calculate the total budget
schoolbudget_column = school_data.iloc[:, 4] 
total_budget = schoolbudget_column.sum()
```

```python
# Calculate the average maths score
studentmaths_score_column = student_data.iloc[:, 6]
average_maths_score = studentmaths_score_column.mean()
```

```python
#Calculate the average reading score
studentreading_score_column = student_data.iloc[:, 5]
average_reading_score = studentreading_score_column.mean()
```

```python
#Calculated the percentage of students who passed both math
passing_math_count = student_data[student_data['maths_score'] >= 50]['maths_score'].count()
percentage_passing_math = (passing_math_count / total_unique_student) * 100
```

```python
#Calculated the percentage of students who passed reading
passing_reading_count = student_data[student_data['reading_score'] >= 50]['reading_score'].count()
percentage_passing_reading = (passing_reading_count / total_unique_student) * 100
```

```python
#Calculated the percentage of students who passed both math and reading (% Overall Passing)
passing_overal_count= student_data[(student_data['maths_score'] >= 50)&(student_data['reading_score'] >= 50)].shape[0]
percentage_overalpassing = (passing_overal_count / total_unique_student) * 100
```

```python
results = {"Total Schools": total_unique_schools,
    "Total Students": total_unique_student,
    "Total Budget":total_budget,
    "Average Maths Score":average_maths_score,
    "Average Reading Score":average_reading_score,
    "% Passing Maths": percentage_passing_math,
    "% Passing Reading": percentage_passing_reading,
    "% Overall Passing": percentage_overalpassing}
# Create a DataFrame from the results dictionary
results_df = pd.DataFrame(results, index=[0],columns=["Total Schools","Total Students","Total Budget","Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"])
#Optional: give the displayed data cleaner formatting
results_df["Total Students"] = results_df["Total Students"].map("{:,.0f}".format)
results_df["Total Budget"] = results_df["Total Budget"].map("${:,.2f}".format)
results_df["% Passing Maths"] = results_df["% Passing Maths"].map("{:.1f}%".format)
results_df["% Passing Reading"] = results_df["% Passing Reading"].map("{:.1f}%".format)
results_df["% Overall Passing"] = results_df["% Overall Passing"].map("{:.1f}%".format)
# Display the DataFrame
results_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths (the percentage of students who passed maths)
  * % Passing Reading (the percentage of students who passed reading)
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Calculate and Create a DataFrame for Total students
student_counts = school_data_complete.groupby(["school_name","type","size"])["student_name"].count().reset_index()
student_counts.rename(columns={"student_name": "Total students"}, inplace=True)
```

```python
# Calculate and Create a DataFrame for total budget
total_budget=school_data_complete.groupby("school_name")["budget"].mean().reset_index()
total_budget.rename(columns={"budget": "Total School Budget"}, inplace=True)
#merge Total students and total budget
school_summary = pd.merge(student_counts,total_budget,how="left", on="school_name")
```

```python
# Calculate and Create a DataFrame for per student budget
per_student_budget=total_budget["Total School Budget"]/student_counts["Total students"]
Per_Student_Budget_df=pd.DataFrame({"school_name": total_budget["school_name"],"Per student budget": per_student_budget})
#merge School summary with per student budget
school_summary = pd.merge(school_summary,Per_Student_Budget_df,how="left", on="school_name")
```

```python
# Calculate and Create a DataFrame for Average maths score
average_maths_score=school_data_complete.groupby("school_name")["maths_score"].mean().reset_index()
average_maths_score.rename(columns={"maths_score": "Average Maths Score"}, inplace=True)
school_summary = pd.merge(school_summary,average_maths_score,how="left", on="school_name")
```

```python
# Calculate and Create a DataFrame for Average reading score
average_reading_score=school_data_complete.groupby("school_name")["reading_score"].mean().reset_index()
average_reading_score.rename(columns={"reading_score": "Average Reading Score"}, inplace=True)
school_summary = pd.merge(school_summary,average_reading_score,how="left", on="school_name")
```

```python
# Calculate and Create a DataFrame for % passing maths (the percentage of students who passed maths)
percentage_students_passed_maths=school_data_complete.groupby("school_name")["maths_score"].apply(lambda x: (x >=50).sum() / len(x) * 100).reset_index()
percentage_students_passed_maths.rename(columns={"maths_score": "% Passing Maths"}, inplace=True)
school_summary = pd.merge(school_summary,percentage_students_passed_maths,how="left", on="school_name")
```

```python
# Calculate and Create a DataFrame for % passing reading (the percentage of students who passed reading)
percentage_students_passed_reading=school_data_complete.groupby("school_name")["reading_score"].apply(lambda x: (x >=50).sum() / len(x) * 100).reset_index()
percentage_students_passed_reading.rename(columns={"reading_score": "% Passing Reading"}, inplace=True)
school_summary = pd.merge(school_summary,percentage_students_passed_reading,how="left", on="school_name")
```

```python
# Calculate and Create a DataFrame for % overall passing (the percentage of students who passed maths AND reading)
percentage_students_passed_overall=school_data_complete.groupby("school_name")[["reading_score", "maths_score"]].apply(lambda x: ((x >= 50).all(axis=1)).sum() / len(x) * 100).reset_index()
percentage_students_passed_overall.columns = ["school_name", "% Overall Passing"]
school_summary = pd.merge(school_summary,percentage_students_passed_overall,how="left", on="school_name")
```

```python
school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per student budget"]=school_summary["Per student budget"].map("${:,.2f}".format)
school_summary2 = school_summary.set_index("school_name")
del school_summary2['size']
school_summary2
#school_summary.dtypes
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Calculate and Create a DataFrame for average maths score for students of each year level (9, 10, 11, 12) at each school
# Create a pandas series for each year. Hint: use a conditional statement.
maths_scores_9 = school_data_complete.loc[school_data_complete['year'] == 9, 'maths_score']
maths_scores_10 = school_data_complete.loc[school_data_complete['year'] == 10, 'maths_score']
maths_scores_11= school_data_complete.loc[school_data_complete['year'] == 11, 'maths_score']
maths_scores_12= school_data_complete.loc[school_data_complete['year'] == 12, 'maths_score']
# Group each series by school
grouped_scores_9=maths_scores_9.groupby(school_data_complete['school_name']).mean(),
grouped_scores_10=maths_scores_10.groupby(school_data_complete['school_name']).mean(),
grouped_scores_11=maths_scores_11.groupby(school_data_complete['school_name']).mean(),
grouped_scores_12=maths_scores_12.groupby(school_data_complete['school_name']).mean(),
# Convert the series into a dataframe
grouped_scores_9_df=pd.DataFrame(grouped_scores_9).T
grouped_scores_10_df=pd.DataFrame(grouped_scores_10).T
grouped_scores_11_df=pd.DataFrame(grouped_scores_11).T
grouped_scores_12_df=pd.DataFrame(grouped_scores_12).T
# Combine the dataframes
maths_scores_year= pd.merge(grouped_scores_9_df, grouped_scores_10_df, how="left", on="school_name")
maths_scores_year.columns=['Year 9','Year 10']
maths_scores_year= pd.merge(maths_scores_year, grouped_scores_11_df, how="left", on="school_name")
maths_scores_year= pd.merge(maths_scores_year, grouped_scores_12_df, how="left", on="school_name")
maths_scores_year.columns=['Year 9','Year 10','Year 11','Year 12']
#give the displayed data cleaner formatting
maths_scores_year["Year 9"] = (maths_scores_year["Year 9"]).map("{:.1f}%".format)
maths_scores_year["Year 10"] = (maths_scores_year["Year 10"]).map("{:.1f}%".format)
maths_scores_year["Year 11"] = (maths_scores_year["Year 11"]).map("{:.1f}%".format)
maths_scores_year["Year 12"] = (maths_scores_year["Year 12"]).map("{:.1f}%".format)
maths_scores_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Select specific columns, sort the DataFrame, and reset the index
school_summary3=school_summary2.iloc[:,[3,4,5,6,7,8]].sort_values("Per student budget",ascending=True).reset_index()

# Remove the dollar sign from "Per student budget" and convert to float
school_summary3["Per student budget"]=school_summary3["Per student budget"].str.replace('$', '',regex=True)
school_summary3["Per student budget"]=school_summary3["Per student budget"].astype(float)
# Create the bins in which Data will be held

bins = [578, 585, 630, 645, 680]

# Create the names for the five bins
group_names = ["<585", "$585-630", "$630-645", "$645-680"]
# Slice the data and place it into bins
school_summary3["Spending Ranges (Per Student)"] = pd.cut(school_summary3["Per student budget"], bins, labels=group_names, include_lowest=True)
# Creating a group based off of the bins
grouped_school_summary=school_summary3.groupby("Spending Ranges (Per Student)")
grouped_school_summary.mean(numeric_only=True).iloc[:,[1,2,3,4]].applymap("{:.2f}".format)
```
